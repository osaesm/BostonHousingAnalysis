```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# from pandas.plotting import scatter_matrix
import seaborn as sns
import mpl_toolkits
%matplotlib inline


f = pd.read_csv('./all/train.csv')
test = pd.read_csv('./all/test.csv')

print(f.head(10))
# print(f.describe())
plt.figure(figsize=(10,10))
sns.jointplot(x = f.age.values, y = f.medv.values, size = 10)
plt.title('crime rate')
plt.xlabel('crime rate')
plt.ylabel('price')
plt.show()
sns.despine
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
reg = LinearRegression()
labels = f['medv']
# y_test = test['medv']
test1 = test.drop(['ID'],axis =1)
train1 = f.drop(['ID', 'medv'],axis = 1)
x_train, x_test, y_train, y_test = train_test_split(train1, labels, test_size = 0.10, random_state = 2)
reg.fit(x_train, y_train)
reg.score(x_test,y_test)
```

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators = 1000, max_depth = 5, min_samples_split = 100, learning_rate = 0.004, loss = 'ls')
clf.fit(x_train, y_train)
clf.score(x_test,y_test)
```

```python
lol = clf.predict(test1)
print(lol)
```

```python
# from sklearn.model_selection import KFold
# kf = 
result = pd.DataFrame(lol,columns = ['medv'])
ids = test['ID']
result = pd.concat([ids,result], axis = 1, sort = False)

print(result)
result.to_csv('result.csv', index = False)
```
